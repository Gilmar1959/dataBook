<h4>Verificando a existência de valores nulos na coluna Id de df_books_rating_cs</h4>

<h3>Por que os dados derivados do dataframe df_books_rating_cs tem tamanho diferentes do df_books_data?</h3>

<p> O df_books_rating_qt_comments tem 221.998 Id's distintos</p>
<p> O df_books_data contem 212404 títulos.

<h4> Resolvi, então, fazer um merge entre os dataframe aplicar a coluna Id do dataframe df_books_rating_cs no dataframe df_books_data_cs, combinando pelo Title, coluna comum a ambos dataframes</h4>

<p>Para isso, selecionei os Id's unicos no dataframe df_books_rating_cs</p>

```python
df_col_Id_Title = df_books_rating_cs.loc[:,['Id', 'Title']]

df_col_Id_Title
```

```python
df_col_Id_Title_SR = df_col_Id_Title.drop_duplicates()

df_col_Id_Title_SR
```

<h4>Incorporando a coluna Id ao dataframe</h4>

```python
df_books = pd.merge(df_books_data_cs, df_col_Id_Title_SR, how = 'left', on = 'Title')

df_books
```

<h4>Obtendo a quantidade de Comentários por livro</h4>

```python
df_qtd_comments_book = df_books_rating_cs['Id'].value_counts().reset_index().rename(columns={"count": "qtd_comments"})

df_qtd_comments_book

```

<h4>Incorporando a quantidade de comentários ao dataframe de livros</h4>

```python
df_books_comments = pd.merge(df_books, df_qtd_comments_book, how = 'left', on = 'Id')

df_books_comments
```

<h4>Listando os maiores valores da coluna qtd_comments</h4>

```python
df_books_comments_20 = df_books_comments.nlargest(20, ['qtd_comments'])
                            
df_books_comments_20
```

```python
df_books_comments_20.plot(kind='bar',x='Title',y='qtd_comments', xlabel='Títulos', ylabel= 'Quantidade') 
graf.xticks(rotation=90)
```

```python
df_books_authors_20 = df_books_comments.nlargest(20, ['qtd_comments'])
                            
df_books_authors_20
```

```python
df_books_authors_20.plot(kind='bar',x='authors',y='qtd_comments', xlabel='Autores', ylabel= 'Quantidade') 
graf.xticks(rotation=90)
```
